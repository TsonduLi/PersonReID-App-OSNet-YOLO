```python
import matplotlib.pyplot as plt
import numpy as np

# Data
fixed_epoch = [10, 20, 30, 35, 50]
pre_transition_accuracy = [90.4883, 90.6250, 95.3125, 95.8464, 93.7500]
post_transition_accuracy = [49.3750, 71.8750, 79.6875, 84.4531, 40.6250]
accuracy_difference = [41.1133, 18.7500, 15.6250, 11.3933, 53.1250]

# Bar plot data
bar_width = 0.35
r1 = np.arange(len(fixed_epoch))

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Create bars with segments
for i in range(len(fixed_epoch)):
    ax1.bar(r1[i], pre_transition_accuracy[i], color='steelblue', edgecolor='grey', width=bar_width)
    ax1.bar(r1[i], post_transition_accuracy[i], color='lightcoral', edgecolor='grey', width=bar_width)

# Adding labels
ax1.set_xlabel('Fixed Epoch')
ax1.set_ylabel('Accuracy')
ax1.set_title('Pre and Post Transition Accuracy with Accuracy Difference')
ax1.set_xticks(r1)
ax1.set_xticklabels(fixed_epoch)

# Adding curve for accuracy difference
ax2 = ax1.twinx()
ax2.plot(r1, accuracy_difference, color='darkgreen', marker='o', label='Accuracy Difference')
ax2.set_ylabel('Accuracy Difference')

# Legend
ax1.legend(['Pre-Transition Accuracy', 'Post-Transition Accuracy'], loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
